Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read CSV

```python
healthcare = pd.read_csv('Healthcare.csv')
```

Count unique medical conditions grouped by gender

```python
unique_medical = healthcare.groupby(['Gender', 'Medical Condition']).size().unstack()
```

Plot data with labels

```python
unique_medical.T.plot(kind='bar', figsize=(12, 8), title='Comparison of Medical Conditions Across Genders')
# Add labels
plt.xlabel('Medical Condition')
plt.ylabel('Number of Cases')
plt.legend(title='Gender')
plt.tight_layout()

# Show the chart
plt.show()
```
